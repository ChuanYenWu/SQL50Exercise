敘述的執行順序
截至目前我們已經學習完所有 SQL 資料查詢語言（Data Query Language, DQL）的保留字，寫作順序必須遵從標準 SQL 的規定，否則執行會發生錯誤，但是寫作順序和關聯式資料庫管理系統的執行順序是不同的，保留字的執行依照下列順序：

1. FROM 與 JOIN
2. WHERE
3. GROUP BY
4. HAVING
5. SELECT
6. DISTINCT
7. UNION
8. ORDER BY
9. LIMIT
暸解敘述的執行順序能夠幫助我們排除一些使用上可能發生的問題，舉例來說，在 WHERE 寫作的條件不能用 SELECT 寫作的縮寫或衍生計算欄位，因為 WHERE 比 SELECT 先執行。

將截至目前所學的 SQL 保留字集中在一個敘述中，寫作順序必須遵從標準 SQL 的規定。

```mysql
SELECT covid19.lookup_table.Country_Region, 
       SUM(covid19.daily_report.Confirmed) AS total_confirmed
  FROM covid19.lookup_table
  JOIN covid19.daily_report
    ON covid19.lookup_table.Combined_Key = covid19.daily_report.Combined_Key
 GROUP BY covid19.lookup_table.Country_Region
 ORDER BY total_confirmed DESC
 LIMIT 10;
```

```mysql
SELECT lookup_table.Country_Region,
       SUM(daily_report.Confirmed) AS total_confirmed
  FROM daily_report
  JOIN lookup_table
    ON daily_report.Combined_Key = lookup_table.Combined_Key
 GROUP BY lookup_table.Country_Region
 ORDER BY total_confirmed DESC
 LIMIT 10;
```

```mysql
SELECT twElection2020.parties.id, 
twElection2020.parties.party, 
CASE WHEN TRUE THEN "不分區立委" END AS election,
ROUND(SUM(twElection2020.legislative_at_large.votes)*1.0 / (SELECT SUM(votes) FROM twElection2020.legislative_at_large) ,4) 
AS votes_percentage

FROM twElection2020.parties
JOIN twElection2020.legislative_at_large
ON twElection2020.parties.id = twElection2020.legislative_at_large.party_id

WHERE twElection2020.parties.party in ("中國國民黨","民主進步黨","親民黨")

GROUP BY twElection2020.parties.party
```

```mysql
SELECT twElection2020.parties.id, 
twElection2020.parties.party, 
"不分區立委" AS election,
ROUND(SUM(twElection2020.legislative_at_large.votes)*1.0 / (SELECT SUM(votes) FROM twElection2020.legislative_at_large) ,4) 
AS votes_percentage

FROM twElection2020.parties
JOIN twElection2020.legislative_at_large
ON twElection2020.parties.id = twElection2020.legislative_at_large.party_id

WHERE twElection2020.parties.party in ("中國國民黨","民主進步黨","親民黨")

GROUP BY twElection2020.parties.party
```

```mysql
SELECT twElection2020.parties.id, 
twElection2020.parties.party, 
"區域立委" AS election,
ROUND(SUM(twElection2020.legislative_regional.votes)*1.0 / (SELECT SUM(votes) FROM twElection2020.legislative_regional) ,4) 
AS votes_percentage


FROM twElection2020.legislative_regional
JOIN twElection2020.candidates
ON twElection2020.legislative_regional.candidate_id = twElection2020.candidates.id
JOIN twElection2020.parties
ON twElection2020.candidates.party_id = twElection2020.parties.id

WHERE twElection2020.parties.party in ("中國國民黨","民主進步黨","親民黨")

GROUP BY twElection2020.parties.party
```

```mysql
SELECT twElection2020.parties.party, 
       "不分區立委" AS election,
       ROUND(
           SUM(twElection2020.legislative_at_large.votes)*1.0 /
                 (SELECT SUM(votes) FROM twElection2020.legislative_at_large)   ,4) AS votes_percentage

  FROM twElection2020.parties
  JOIN twElection2020.legislative_at_large
    ON twElection2020.parties.id = twElection2020.legislative_at_large.party_id
 WHERE twElection2020.parties.party in ("中國國民黨","民主進步黨","親民黨")
 GROUP BY twElection2020.parties.party
 
UNION

SELECT twElection2020.parties.party, 
       "區域立委" AS election,
       ROUND(
           SUM(twElection2020.legislative_regional.votes)*1.0 /
               (SELECT SUM(votes) FROM twElection2020.legislative_regional)   ,4) AS votes_percentage

  FROM twElection2020.legislative_regional
  JOIN twElection2020.candidates
    ON twElection2020.legislative_regional.candidate_id = twElection2020.candidates.id
  JOIN twElection2020.parties
    ON twElection2020.candidates.party_id = twElection2020.parties.id
 WHERE twElection2020.parties.party in ("中國國民黨","民主進步黨","親民黨")
 GROUP BY twElection2020.parties.party
 
 ORDER BY election;
```

```mysql
SELECT parties.party,
       '不分區立委' AS election,
       ROUND(CAST(SUM(legislative_at_large.votes) AS REAL) / (SELECT SUM(legislative_at_large.votes) FROM legislative_at_large), 4) AS votes_percentage
  FROM legislative_at_large
  JOIN parties
    ON legislative_at_large.party_id = parties.id
 WHERE parties.party IN ('中國國民黨', '民主進步黨', '親民黨')
 GROUP BY legislative_at_large.party_id
 UNION
SELECT parties.party,
       '區域立委' AS election,
       ROUND(CAST(SUM(legislative_regional.votes) AS REAL) / (SELECT SUM(legislative_regional.votes) FROM legislative_regional), 4) AS votes_percentage
  FROM legislative_regional
  JOIN candidates
    ON legislative_regional.candidate_id = candidates.id
  JOIN parties
    ON candidates.party_id = parties.id
 WHERE parties.party IN ('中國國民黨', '民主進步黨', '親民黨')
 GROUP BY parties.id
 ORDER BY election;
```

```mysql
SELECT nba.teams.fullName AS team_name,
       nba.players.firstName || ' ' || nba.players.lastName AS player_name,
       nba.career_summaries.ppg
  FROM nba.career_summaries
  JOIN nba.players
    ON nba.career_summaries.personId = nba.players.personId
  JOIN nba.teams
    ON nba.players.teamId = nba.teams.teamId

 WHERE nba.teams.fullName = 'Los Angeles Lakers'
 ORDER BY nba.career_summaries.ppg DESC;
```

```mysql
SELECT teams.fullName AS team_name,
       players.firstName || ' ' || players.lastName AS player_name,
       career_summaries.ppg
  FROM players
  JOIN teams
    ON players.teamId = teams.teamId
  JOIN career_summaries
    ON players.personId = career_summaries.personId
 WHERE teams.nickname = 'Lakers'
 ORDER BY career_summaries.ppg DESC;
```

```mysql
SELECT nba.teams.fullName AS team,
       nba.players.firstName || ' ' || nba.players.lastName AS player,
       MAX(nba.career_summaries.ppg) AS ppg
  FROM nba.career_summaries
  JOIN nba.players
    ON nba.career_summaries.personId = nba.players.personId
  JOIN nba.teams
    ON nba.players.teamId = nba.teams.teamId

 GROUP BY team;
```

```mysql
SELECT teams.fullName AS team,
       players.firstName || ' ' || players.lastName AS player,
       MAX(career_summaries.ppg) AS ppg
  FROM players
  JOIN teams
    ON players.teamId = teams.teamId
  JOIN career_summaries
    ON players.personId = career_summaries.personId
 GROUP BY teams.fullName
 ORDER BY teams.fullName;
```

```mysql
SELECT movies.release_year,
       movies.title,
       actors.name,
       CASE WHEN casting.ord=1 THEN 1
            ELSE 0 END AS is_lead_actor
  FROM casting
  JOIN movies
    ON casting.movie_id = movies.id
  JOIN actors
    ON casting.actor_id = actors.id
 WHERE actors.name IN ( 'Tom Hanks', 'Leonardo DiCaprio' )
 ORDER BY movies.release_year;
```

```mysql
SELECT movies.release_year,
       movies.title,
       actors.name,
       casting.ord = 1 AS is_lead_actor
  FROM actors
  JOIN casting
    ON actors.id = casting.actor_id
  JOIN movies
    ON casting.movie_id = movies.id
 WHERE actors.name IN ('Tom Hanks', 'Leonardo DiCaprio')
 ORDER BY movies.release_year;
```
