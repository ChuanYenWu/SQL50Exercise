```mysql
SELECT *
  FROM movies
 ORDER BY release_year DESC, -- descending
          rating,            -- ascending
          title              -- ascending
 LIMIT 10;
```

```mysql
SELECT title
  FROM movies
 ORDER BY release_year DESC, -- descending
          rating            -- ascending
 LIMIT 10;
```

```mysql
SELECT runtime,
       runtime / 60 AS hours,
       runtime % 60 AS minutes
  FROM movies
 ORDER BY hours DESC,
          minutes
 LIMIT 10;
```

```mysql
SELECT *
  FROM movies
 ORDER BY runtime
 LIMIT 10;
```

```mysql
SELECT *
  FROM movies
 ORDER BY runtime DESC
 LIMIT 10;
```
