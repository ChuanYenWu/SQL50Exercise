將截至目前所學的資料查詢語言（DQL）集中在一個敘述中，寫作順序必須遵從標準 SQL 的規定。
SELECT DISTINCT columns AS alias,
       CASE WHEN condition_1 THEN result_1
            WHEN condition_2 THEN result_2
            ...
            ELSE result_n END AS alias
  FROM table
 WHERE conditions
 GROUP BY columns
HAVING conditions
 ORDER BY columns DESC
 LIMIT m;

```mysql
SELECT Country_Region, SUM(Confirmed) AS total_confirmed
FROM covid19.daily_report
JOIN covid19.lookup_table
ON covid19.daily_report.Combined_Key = covid19.lookup_table.Combined_Key
GROUP BY Country_Region
LIMIT 5;
```

```mysql
CREATE VIEW covid19.total_confirmed_by_country_region
    AS
SELECT Country_Region, SUM(Confirmed) AS total_confirmed
  FROM covid19.daily_report
  JOIN covid19.lookup_table
    ON covid19.daily_report.Combined_Key = covid19.lookup_table.Combined_Key
 GROUP BY Country_Region;
```

```mysql
SELECT *
  FROM covid19.total_confirmed_by_country_region
 LIMIT 5;
```

```mysql
CREATE VIEW total_confirmed_by_country_region
    AS 
SELECT lookup_table.Country_Region,
       SUM(daily_report.Confirmed) AS total_confirmed
  FROM daily_report
  JOIN lookup_table
    ON daily_report.Combined_Key = lookup_table.Combined_Key
 GROUP BY lookup_table.Country_Region;
```

```mysql
SELECT *
  FROM total_confirmed_by_country_region;
```

```mysql
CREATE VIEW nba.ppg_leader_by_teams
    AS 

SELECT nba.teams.fullName AS team,
       nba.players.firstName,
       nba.players.lastName,
       MAX(nba.career_summaries.ppg) AS ppg
  FROM nba.career_summaries
  JOIN nba.players
    ON nba.career_summaries.personId = nba.players.personId
  JOIN nba.teams
    ON nba.players.teamId = nba.teams.teamId

 GROUP BY team;
```

```mysql
SELECT *
  FROM nba.ppg_leader_by_teams;
```

```mysql
DROP VIEW nba.ppg_leader_by_teams
```
