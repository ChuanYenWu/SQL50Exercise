```mysql
SELECT firstName,
       lastName,
       heightMeters
  FROM nba.players
 WHERE heightMeters IN (
                        SELECT MIN(heightMeters),
                               MAX(heightMeters)
                          FROM nba.players
);
```

```mysql
SELECT firstName,
       lastName,
       heightMeters
  FROM nba.players
 WHERE (
         heightMeters = (SELECT MIN(heightMeters) FROM nba.players) OR  
         heightMeters = (SELECT MAX(heightMeters) FROM nba.players)
);
```

```mysql
SELECT firstName,
       lastName,
       heightMeters
  FROM players
 WHERE heightMeters = (SELECT MAX(heightMeters) FROM players) OR
       heightMeters = (SELECT MIN(heightMeters) FROM players);
```

```mysql
SELECT COUNT( * ) AS count_player_country
             FROM nba.players
            GROUP BY country
```

```mysql
SELECT country,
(
           SELECT COUNT( * ) AS count_player_country
             FROM nba.players
            GROUP BY country
       )* 1.0 / (SELECT COUNT(*)   FROM players) AS player_percentage
  FROM nba.players;
```

The SQL query you’re trying to execute seems to be calculating the percentage of players from each country. However, it appears to have a syntax error because you’re trying to use a subquery that returns multiple rows (due to the GROUP BY clause) in a context where only a single value is expected.

```mysql
SELECT country,
       COUNT(*) * 1.0 / (SELECT COUNT(*) FROM nba.players) AS player_percentage
  FROM nba.players
 GROUP BY country;
```

```mysql
SELECT country,
       CAST(COUNT(*) AS REAL) / (SELECT COUNT(*) 
                                   FROM players) AS player_percentage
  FROM players
 GROUP BY country
 ORDER BY player_percentage DESC,
          country;
```
